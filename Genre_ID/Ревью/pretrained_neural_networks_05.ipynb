```python
from keras.applications.resnet50 import ResNet50, preprocess_input, decode_predictions
from keras.applications.vgg16 import VGG16, preprocess_input, decode_predictions
from keras.applications.inception_v3 import InceptionV3, preprocess_input, decode_predictions
from keras.applications.densenet import DenseNet121, preprocess_input, decode_predictions
from keras.applications.mobilenet_v2 import MobileNetV2, preprocess_input, decode_predictions
from keras.applications.xception import Xception, preprocess_input, decode_predictions
from keras.preprocessing import image

import numpy as np
import pandas as pd

import os
```

```python

# Загрузка предварительно обученной модели ResNet50
model = ResNet50(weights='imagenet')

# Функция классификации жанра по изображению обложки музыкального диска
def classify_genre(image_path):
    img = image.load_img(image_path, target_size=(224, 224))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)

    preds = model.predict(x, verbose=0)
    decoded_preds = decode_predictions(preds, top=3)[0]  # Получаем топ-3 предсказания
    return decoded_preds

# Пример использования
image_path = 'D:\\Data_Science\\Genre_ID\\data\\pop\\0bdfa7be-f306-4173-bf97-edb990625870.png'
predictions = classify_genre(image_path)
for pred in predictions:
    print(f"Жанр: {pred[1]}, Вероятность: {pred[2]}")
```

```python
# Загрузка предварительно обученной модели ResNet50
model = ResNet50(weights='imagenet')

# Путь к каталогу с изображениями
data_dir = 'data'

# Функция для получения изображений и меток из всех подкаталогов
def load_images_and_labels(data_dir):
    images = []
    labels = []
    for root, dirs, files in os.walk(data_dir):
        for subdir in dirs:
            subdir_path = os.path.join(root, subdir)
            subdir_images = [os.path.join(subdir_path, file) for file in os.listdir(subdir_path) if file.endswith('.png')]
            images.extend(subdir_images[:10])  # Ограничение до 10 изображений для каждого подкаталога
            labels.extend([subdir] * min(len(subdir_images), 10))  # Добавление меток для каждого изображения
    return images, labels

# Загрузка изображений и меток из всех подкаталогов
images, true_labels = load_images_and_labels(data_dir)

# Функция для выполнения предсказаний для каждого изображения
def predict_genre(image_path):
    img = image.load_img(image_path, target_size=(224, 224))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)

    preds = model.predict(x, verbose=0)
    decoded_preds = decode_predictions(preds, top=1)[0]  # Получаем одно лучшее предсказание
    return decoded_preds[0][1]

# Выполнение предсказаний для всех изображений
predictions = [predict_genre(image_path) for image_path in images]

# Создание DataFrame из результатов предсказаний и истинных меток
df_resnet50 = pd.DataFrame({'Предсказание': predictions, 'Название каталога': true_labels})

# Вывод DataFrame
print(df_resnet50)
```
